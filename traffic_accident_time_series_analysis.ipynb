#### <a id='toc1_1_1_1_'></a>[Có tháng nào có số vụ tai nạn đặt biệt cao so với kỳ vọng (trung bình) có ý nghĩa thống kê (sử dụng kiểm định Chi-square goodness of fit)](#toc0_)

```python
query_1 = """
    SELECT 
        DATE,
        COUNT([ACCIDENT ID]) AS accidents
    FROM traffic_accidents
    GROUP BY DATE
    ORDER BY DATE
"""

df_anal1 = pd.read_sql(query_1,conn)
df_anal1['month'] = df_anal1['DATE'].dt.month
df_anal1['weekday'] = df_anal1['DATE'].dt.weekday # (0 = Monday)
```

```python
#2023
# sử dụng phương pháp Chi-Square goodness of fit kiểm tra xem 
accidents_2023 = df_anal1.loc[df_anal1['DATE'].dt.year == 2023,['month','accidents']]
monthly_count = accidents_2023.groupby('month')['accidents'].sum()

obs = monthly_count.values
exp_value = np.full_like(obs,obs.mean())

stats , pvalue = chisquare(f_obs=obs,f_exp=exp_value)

if pvalue > 0.05:
    print(f'pvalue = {pvalue} --> Trong năm 2023, Số vụ tai nạn phân bố không khác biệt đáng kể so với phân phối đồng đều quanh trung bình')
else:
    print(f'pvalue = {pvalue} --> Trong năm 2023 có ít nhất 1 tháng có phân phối số vụ tai nạn khác biệt đáng kể')
    std_resid = (obs - exp_value) / np.sqrt(exp_value) # tính phần dư chuẩn hóa
    std_resid_df = pd.DataFrame({
        'Month':range(1,13),
        'Observation':obs,
        'Mean':exp_value,
        'Standardized Residual':std_resid
    })
    std_resid_df['Significant'] = std_resid_df['Standardized Residual'].apply(lambda x: 'yes' if abs(x) > 1.96 else 'no')
    print(std_resid_df)
pvalue
```

```python
#2024
# sử dụng phương pháp Chi-Square goodness of fit kiểm tra xem 
accidents_2024 = df_anal1.loc[df_anal1['DATE'].dt.year == 2024,['month','accidents']]
monthly_count = accidents_2024.groupby('month')['accidents'].sum()

obs = monthly_count.values
exp_value = np.full_like(obs,obs.mean())
exp_value = exp_value * (obs.sum() / exp_value.sum())  # Điều chỉnh tổng

stats , pvalue = chisquare(f_obs=obs,f_exp=exp_value)

if pvalue > 0.05:
    print(f'pvalue = {pvalue} --> Trong năm 2024, Số vụ tai nạn phân bố không khác biệt đáng kể so với phân phối đồng đều quanh trung bình ')
else:
    print(f'pvalue = {pvalue} --> Trong năm 2024 có ít nhất 1 tháng có phân phối số vụ tai nạn khác biệt đáng kể')
    std_resid = (obs - exp_value) / np.sqrt(exp_value) # tính phần dư chuẩn hóa
    std_resid_df = pd.DataFrame({
        'Month':range(1,13),
        'Observation':obs,
        'Mean':exp_value,
        'Standardized Residual':std_resid
    })
    std_resid_df['Significant'] = std_resid_df['Standardized Residual'].apply(lambda x: 'yes' if abs(x) > 1.96 else 'no')
    print(std_resid_df)
```
